```python
contagem_desconto = df_amazon['discounted_price'].value_counts()
contagem_desconto.plot(kind = 'bar', edgecolor = 'blue', color = 'pink')

plt.xlabel('Desconto')
plt.ylabel('Quantidade')
plt.title('Quantidade de desconto')
```

```python
contagem_categorias = df_amazon['category'].value_counts()
contagem_categorias.plot(kind = 'bar', edgecolor = 'blue', color = 'pink')

plt.xlabel('Categorias')
plt.ylabel('Quantidade')
plt.title('Quantidade por categoria')
```
