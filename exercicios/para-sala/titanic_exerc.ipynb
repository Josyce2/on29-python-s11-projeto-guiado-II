```python
def processamento_df(df_titanic):
    df_titanic.head(3)

    baby_df = df_titanic.sample(100)

    df_titanic =df_titanic.sort_values('Fare')

    df_titanic = df_titanic.drop(['SibSp', 'Parch'], axis = 'columns')

    df_titanic = df_titanic.dropna(subset = 'Fare')

    df_titanic = df_titanic.reset_index()

    metricas = df_titanic.describe()

    return baby_df, df_titanic, metricas
```

```python
#leitura csv
df_titanic = pd.read_csv('titanic.csv')
```

```python
baby_df, df_titanic, metricas = processamento_df(df_titanic)
```

```python
contagem_passageiros = df_titanic['Pclass'].value_counts()
contagem_passageiros.plot(kind = 'bar', edgecolor = 'blue', color = 'pink')

plt.xlabel('Classe do passageiro')
plt.ylabel('Quantidade')
plt.title('Passageiros por classe')
```

```python
#taxa sobrevivencia por sexo
taxa_sob_sexo = df_titanic.groupby('Sex')['Survived'].mean()

cores = ['lightpink', 'lightblue']

barras = taxa_sob_sexo.plot.bar(edgecolor = 'blue', color = cores)

plt.xlabel('Sexo')
plt.ylabel('Taxa de Sobrevivencia')
plt.title('Taxa de sobrevivencia por Sexo')

plt.legend(['Taxa de Sobrevivencia'])

for i, v in enumerate(taxa_sob_sexo):
    barras.text(i, v + 0.01, f'{v:2f}', color='black', ha = 'center')
```

```python
# distribuicao de idade
plt.hist(df_titanic['Age'], bins = 10, edgecolor = 'blue', color = 'pink')

plt.xlabel('Idade')
plt.ylabel('Quantidade')
plt.title('Distribuicao de idade')

plt.legend(['Quantidade'])

for ret in plt.gca().patches:
    altura = ret.get_height()
    plt.gca().annotate(f'{int(altura)}',xy=(ret.get_x() + ret.get_width()/ 2, altura),
                      xytext=(0,5), textcoords='offset points', ha = 'center', va = 'bottom')
```

```python
def tratar_dados_titanic(df_titanic):

    df_titanic.dropna(subset=['Cabin'], inplace=True)

    df_titanic['Age'].replace([np.inf, -np.inf], np.nan, inplace =True)
    df_titanic['Age'].fillna(-1, inplace = True)

    df_titanic['Age'] = df_titanic['Age'].astype(int)

    return df_titanic

```

```python
df_titanic_tratado = tratar_dados_titanic(df_titanic)
```
